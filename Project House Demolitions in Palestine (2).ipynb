# Convert Data Type

### = Convert data type to number in columns that represent numbers

```python
columns_to_convert=['Housing Units','People Left Homeless','Minors Left Homeless']
for col in columns_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')  #    

print(df)
```

### = Convert columns to dates with default start of year

```python
df['Year'] = pd.to_datetime(df['Year'], format='%Y') 
```

# missing values:

### * Empty values in the 'Demolition carried out by' column represent the Israeli occupation.

```python
df['Demolition carried out by'] = df['Demolition carried out by'].fillna("Israeli occupation")
```

### * Fill all NaN with 0 in People Left Homeless and Minors Left Homeless


```python
df[['People Left Homeless', 'Minors Left Homeless']] = df[['People Left Homeless', 'Minors Left Homeless']].fillna(0)
```

# cleaning data:

### - Remove duplicates function df.drop_duplicates()

```python
df = df.drop_duplicates()
```

### - Remove the line that does not contain data with function df.drop()

```python
df.drop(269, inplace=True)#              
```

### Step 4: Decision Making  

**Conclusion**  

This report analyzed data on forced house demolitions in Palestine from 2004 to 2021 and their impact on the displacement of Palestinian citizens. The analysis focused on testing two key statistical hypotheses:  

1. **The number of houses demolished annually is increasing due to the rise in settlement projects.**  
2. **The frequency of demolition processes in East Jerusalem is higher, aiming to pressure residents and force them to leave.**  

### Key Findings:  

1. **First Hypothesis: Rejected**  
   - The analysis disproved the first hypothesis, as the highest number of house demolitions occurred in 2004, and no clear trend of annual increase was observed over the years analyzed.

2. **Second Hypothesis: Supported**  
   - The second hypothesis was confirmed, as East Jerusalem ranked first in terms of the frequency of demolition decisions, indicating a pattern of intensified pressure on residents in this area.

### Methodology:  

#### 1. **Pre-Processing and Data Cleaning**  

- **Data Formatting**: All date variables were converted to proper date and time formats for consistency.  
- **Handling Missing Values**: Missing values were addressed where applicable, except for variables like partial or complete demolition, as these could not be accurately predicted.  
- **Duplicate Removal**: Data was reviewed for duplicates or irregularities in user logs, and no significant issues were found.  

#### 2. **Descriptive Statistics**  

- Descriptive statistics were conducted to understand the dataset comprehensively, including measures of central tendency and variance.  
- The analysis revealed significant disparities in the number of people left homeless per demolition case, highlighting the varied impact of these events.  

#### 3. **Exploratory Analysis**  

- The dataset was explored in depth to uncover patterns and trends in house demolitions and their consequences.  
- A specific focus was placed on analyzing the behavior and frequency of demolitions across regions, with comparisons between the West Bank, East Jerusalem, and the Gaza Strip.  

---

### Summary of Findings:  

1. **Disparity in Displacement**:  
   - There is a notable variation in the number of people left homeless per demolition case, suggesting that some incidents have a disproportionately larger impact on displacement.  

2. **Regional Focus**:

3. **Temporal Patterns**:  
   - Despite the ongoing settlement projects, the data did not support the assumption of a consistent annual increase in demolitions.
